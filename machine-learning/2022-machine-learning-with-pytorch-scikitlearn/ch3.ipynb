```python
import numpy as np
from sklearn import datasets

dataset = datasets.load_iris()
X = dataset.data[:, [2, 3]]
y = dataset.target
np.bincount(np.where(y == 0, 5, 0))
```

```python
import time

import matplotlib.pyplot as plt
import psutil
import seaborn as sns

tracking_duration = 60 * 4  # in seconds
tracking_interval = 1  # in seconds
process_name = "Solr"
PID = 60465  # process id

sns.set_theme()
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.dpi"] = 200

start = time.time()
cpu_utilization = []

p = psutil.Process(PID)

while (time.time() - start) < tracking_duration:
    cpu_utilization.append(p.cpu_percent(interval=tracking_interval))

plt.plot(range(len(cpu_utilization)), cpu_utilization)
plt.title(f"{process_name} Cpu Utilization")
plt.xlabel("Time (seconds)")
plt.ylabel("CPu Usage (%)")
plt.show()
```
